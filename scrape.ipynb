```python
from google_play_scraper import reviews
import pandas as pd
import time

all_reviews = []
token = None
total_reviews = 400
batch_size = 200  # max per fetch

while len(all_reviews) < total_reviews:
    count = min(batch_size, total_reviews - len(all_reviews))
    result, token = reviews(
        'com.blitz.blitzandapp',
        lang='id',  # Bahasa Indonesia
        country='id',  # Play Store Indonesia
        count=count,
        continuation_token=token
    )
    
    all_reviews.extend(result)
    print(f"Fetched {len(all_reviews)} / {total_reviews} reviews...")

    # Tambahan delay biar aman
    time.sleep(1)

# Konversi ke DataFrame
df = pd.DataFrame(all_reviews)

# Simpan ke CSV kalau mau
df.to_csv('blitz_reviews_10000.csv', index=False)

print("✅ Selesai! Data disimpan di blitz_reviews_10000.csv")
```

```python
from google_play_scraper import reviews
import pandas as pd
import time

# Pengaturan
package_name = 'com.blitz.blitzandapp'
batch_size = 30
save_every = 100
total_target = 1000

all_reviews = []
token = None
batch_num = 1

while len(all_reviews) < total_target:
    count = min(batch_size, total_target - len(all_reviews))
    
    print(f"\n🔄 Batch {batch_num}: Fetching {count} reviews...")
    start_time = time.time()

    try:
        result, token = reviews(
            package_name,
            lang='id',
            country='id',
            count=count,
            continuation_token=token
        )
    except Exception as e:
        print(f"❌ Error saat mengambil batch: {e}")
        print("⏸️ Istirahat 5 detik, lalu lanjut...")
        time.sleep(5)
        continue

    all_reviews.extend(result)
    batch_time = time.time() - start_time
    print(f"✅ Batch {batch_num} selesai: +{len(result)} review | Total: {len(all_reviews)} | Waktu: {batch_time:.2f} detik")

    # Simpan sementara setiap 500 review
    if len(all_reviews) % save_every == 0 or len(all_reviews) == total_target:
        temp_df = pd.DataFrame(all_reviews)
        temp_df.to_csv(f"review_simpan_{len(all_reviews)}.csv", index=False)
        print(f"💾 Data sementara disimpan ke: review_simpan_{len(all_reviews)}.csv")

    time.sleep(1)  # Delay aman antar batch
    batch_num += 1

# Simpan final
df = pd.DataFrame(all_reviews)
df.to_csv('blitz_reviews_10000_final.csv', index=False)
print("\n✅ Selesai! Semua review disimpan di blitz_reviews_10000_final.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager

options = Options()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

app_id = 'com.blitz.blitzandapp'
driver.get(f'https://play.google.com/store/apps/details?id={app_id}1&hl=id')

print("Silakan klik tombol 'Lihat semua ulasan' secara manual di Chrome...")
input("⏳ekan ENTER jika modal ulasan sudah terbuka dan siap di-scrape...")

try:
    scrollable_div = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, '//div[@role="dialog"]//div[contains(@class, "fysCi")]'))
    )
except:
    print("Gagal menemukan area scroll dalam modal.")
    driver.quit()
    exit()

scroll_times = 500
for i in range(scroll_times):
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', scrollable_div)
    print(f"🔄 Scrolling modal... ({i+1}/{scroll_times})")
    time.sleep(2)

print("⏳ Menunggu review tampil penuh...")
time.sleep(5)

review_elements = driver.find_elements(By.XPATH, '//div[@role="article"]')
if len(review_elements) == 0:
    print("⚠️ Tidak ditemukan review dengan role='article', mencoba alternatif...")
    review_elements = driver.find_elements(By.XPATH, '//div[contains(@class,"RHo1pe")]')

print(f"Jumlah review element terdeteksi: {len(review_elements)}")

reviews_data = []
for idx, r in enumerate(review_elements):
    try:
        nama = r.find_element(By.CLASS_NAME, 'X5PpBb').text
        tanggal = r.find_element(By.CLASS_NAME, 'bp9Aid').text
        komentar = r.find_element(By.CLASS_NAME, 'h3YV2d').text
        rating_div = r.find_element(By.XPATH, './/div[@role="img"]')
        rating = rating_div.get_attribute('aria-label')

        print(f"✅ Review {idx+1}: {nama} | {tanggal} | {rating}")
        reviews_data.append({
            'nama': nama,
            'tanggal': tanggal,
            'rating': rating,
            'komentar': komentar
        })
    except Exception as e:
        print("Gagal ambil review:", e)

df = pd.DataFrame(reviews_data)
df.to_csv('blitz_reviews_fix1.csv', index=False, encoding='utf-8-sig')
print(f" {len(df)}")
driver.quit()
```
